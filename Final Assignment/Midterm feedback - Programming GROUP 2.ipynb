# Introduction: Flight movements for each Dutch airport    (2012-2021)

The aviation sector has developed enormously the last decade. 
Before effects can be determined it is crucial to indicate the flight movements during the last decade for each Dutch airport. From the result the increase or decrease ratio can be determined in order to seek which airport grew the most. 

**What is the amount of flights operated for each aiport?**

A pie chart visualisation is choosen to answer the first subquestion. *What is the ratio between the different aiports in The Netherlands?* Four pie chart are shown at four different periods (i.e 1997, 2002, 2010, 2021). 1997 is selected as it is the start of the dataset. 2002 is just after the twin towers disaster. 2010 is during the financial crisis. 2021 is the most recent data of our dataset. 

The biggest part of the flights from and to The Netherlands can be found at Schiphol Airport, with around 85 to 90 percent.
The other four Dutch airports take the remaining 10 to 15 percent. During the years, Eindhoven Aiport overtook 
Rotterdam The Hague Airport for the position as the second-most used airport in the Netherlands. The percentage for Groningen
Eelde airport has been decreasing overtime.

In summury, from these pie charts there is no significant shift in ratio between total flights over the different Airports in the Netherlands

```python
#filtering on only years in the dataset
filter = do['Periods'].str.contains('JJ', case=False, na=True)
dh = do[filter]

#filtering out the Total Dutch Airports
filter3 = dh['Airports'].isin(['Groningen Eelde','AMS Schiphol','Eindhoven','Maastricht Aachen','Rotterdam The Hague'])
dc = dh[filter3]

#filtering on four different years
filter4a = dc['Periods'].isin(['1997JJ00'])
di = dc[filter4a]
filter4b = dc['Periods'].isin(['2002JJ00'])
dj = dc[filter4b]
filter4c = dc['Periods'].isin(['2010JJ00'])
dk = dc[filter4c]
filter4d = dc['Periods'].isin(['2021JJ00'])
dl = dc[filter4d]


#creating four different pie chart based on different years
figa = px.pie(di, values='TotalFlights_3', names='Airports', title='Ratio between the different aiports in The Netherlands in 1997', color='Airports',
             color_discrete_map={'AMS Schiphol':'lightcyan',
                                 'Rotterdam The Hague':'red',
                                 'Eindhoven':'green',
                                 'Maastricht Aachen':'pink',
                                 'Groningen Eelde': 'orange'})
figa.show()

figb = px.pie(dj, values='TotalFlights_3', names='Airports', title='Ratio between the different aiports in The Netherlands in 2002', color='Airports',
             color_discrete_map={'AMS Schiphol':'lightcyan',
                                 'Rotterdam The Hague':'red',
                                 'Eindhoven':'green',
                                 'Maastricht Aachen':'pink',
                                 'Groningen Eelde': 'orange'})
figb.show()

figc = px.pie(dk, values='TotalFlights_3', names='Airports', title='Ratio between the different aiports in The Netherlands in 2010', color='Airports',
             color_discrete_map={'AMS Schiphol':'lightcyan',
                                 'Rotterdam The Hague':'red',
                                 'Eindhoven':'green',
                                 'Maastricht Aachen':'pink',
                                 'Groningen Eelde': 'orange'})
figc.show()

figd = px.pie(dl, values='TotalFlights_3', names='Airports', title='Ratio between the different aiports in The Netherlands in 2021', color='Airports',
             color_discrete_map={'AMS Schiphol':'lightcyan',
                                 'Rotterdam The Hague':'red',
                                 'Eindhoven':'green',
                                 'Maastricht Aachen':'pink',
                                 'Groningen Eelde': 'orange'})
figd.show()
```

# Development of CO2 emissions for restricted airports

Nowadays, global warming is an actual topic. Therefore, it is interesting to find out the effect of the amount of flights on the CO2-emissions.  

**What are the effects of in- or decreased amount of flights on the CO2-emissions?**

The graph below shows the total flights as a line and the CO2-emission as a histogram over the period 2012-2021 in the Netherlands. Comparing the CO2 emissions of the periods 2015 and 2016 gives a clear difference in the graph. The CO2-emission in 2016 is visibly higher than in 2015, but also the line related to the total flights is rising. In the period between 2016 and 2019 the amount of CO2-emission stays alomst constant. This observation holds for the total flights aswell. The graph show that in 2020 the CO2-emission and the total number of flights will drop abruptly, and this can be related to the corona virus. 

In conclusion, the total flights and the CO2-emission move synchronously. This is a logical outcome, because more flights will cause more CO2-emission. 

In addition, the total flights in 2021 have rised in comparison to 2020, but this does not hold for the CO2-emission. This can be explained by means of the remark on the website of CBS: "Preliminary figures". The given data has not been confirmed, it still could change, and thus it could be higher like the total flights.
