# Matplotlib

## A simple plot

$y=sin(x)$

```python
x = np.linspace(0, 2*math.pi, 100)
y = scipy.sin(x)
plt.plot(x,y)
# plt.savefig('line_plot.png', dpi=200)
plt.show()
```

## Title

```python
x = np.linspace(0, 2*math.pi, 10)
y = scipy.sin(x)
fig, ax = plt.subplots()               #important! this is the proper way!
ax.plot(x,y)
ax.set_title('y = sin(x)')
plt.show()
```

## Axes

```python
x = np.linspace(0, 2*math.pi, 10)
y = scipy.sin(x)
fig, ax = plt.subplots()
ax.plot(x,y)
# set title
ax.set_title('y = sin(x)')
# horizontal axis
ax.set_xlabel('x')
# vertical axis
ax.set_ylabel('y')
plt.show()
```

## Tick

Example from a [matplotlib website](https://matplotlib.org/stable/gallery/ticks_and_spines/ticklabels_rotation.html#sphx-glr-gallery-ticks-and-spines-ticklabels-rotation-py)

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 6]
labels = ['Frogs', 'Hogs', 'Bogs', 'Slogs']

plt.plot(x, y)
# You can specify a rotation for the tick labels in degrees or with keywords.
plt.xticks(x, labels, rotation='vertical')
# Pad margins so that markers don't get clipped by the axes
plt.margins(0.2)
# Tweak spacing to prevent clipping of tick-labels
plt.subplots_adjust(bottom=0.15)
plt.show()
```

## Line properties

For a complete list of line properties, visit this [link](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

An example of line styles are shown bellow:

<img src="line_styles.png" alt="drawing" width="800"/>

[Source](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html)



```python
x = np.linspace(0, 2*math.pi, 10)
y = scipy.sin(x)
fig, ax = plt.subplots()
ax.plot(
    x,y, 
    linestyle='dashdot', linewidth=3, 
    marker='o', markerfacecolor='r'
)
# set title
ax.set_title('y = sin(x)')
# horizontal axis
ax.set_xlabel('x')
# vertical axis
ax.set_ylabel('y')
plt.savefig('line_properties_example.png', dpi=200)
plt.show()
```

## Scatterplot

[iris](https://archive.ics.uci.edu/ml/datasets/iris) dataset: multiple species with following information
- Sepal & petal
- Length and width

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris)
plt.show()
```

## Challenge II

## Relplot

```python
sns.relplot(x='sepal_length', y='sepal_width', data=iris, hue='species')
plt.show()
```

```python
sns.relplot(x='petal_length', y='petal_width', data=iris, hue='species')
# plt.savefig('iris_petal.png', dpi=300)
plt.show()
```

```python
fmri = sns.load_dataset("fmri")
sns.relplot(data=fmri, x="timepoint", y="signal", kind="line")
```

## Jointplot

[tips](https://rdrr.io/cran/regclass/man/TIPS.html) dataset

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
# plt.savefig('total_bill_tip.png', dpi=200)
plt.show()
```

## Boxplot / Violinplot

```python
fig, ax = plt.subplots(ncols=2, figsize=(15,9))
sns.boxplot(x='species', y='petal_width', data=iris, ax=ax[0])
sns.violinplot(x='species', y='petal_width', data=iris, ax=ax[1])
# fig.savefig('iris_petal_width_dist.png', dpi=200)
plt.show()
```

## FacetGrid

```python
g = sns.FacetGrid(tips, col='sex', hue='smoker')
g.map(sns.scatterplot, 'total_bill', 'tip')
g.add_legend()
plt.savefig('smoker.png', dpi=200)
plt.show()
```

## Heatmap

```python
conf_mat = np.array([
    [85, 3, 2, 10, 0],
    [1, 80, 10, 1, 8],
    [0, 0, 70, 20, 10],
    [5, 2, 3, 75, 15],
    [10, 0, 0, 0, 90]
])
sns.heatmap(conf_mat, cmap='YlGn', annot=True)

# fix for mpl bug that cuts off top/bottom of seaborn viz
# https://github.com/mwaskom/seaborn/issues/1773
b, t = plt.ylim() # discover the values for bottom and top
b += 0.5 # Add 0.5 to the bottom
t -= 0.5 # Subtract 0.5 from the top
plt.ylim(b, t) # update the ylim(bottom, top) values

# plt.savefig('heatmap.png', dpi=200)
plt.show()
```
